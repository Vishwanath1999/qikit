```python
from qiskit import *
from math import pi, sin, cos
import numpy as np
from qiskit.visualization import *
from matplotlib import style
style.use('dark_background')
```

```python
def plot_bloch_vector_sphere(theta, phi, r):
    x = r*sin(theta)*cos(phi)
    y = r*sin(theta)*sin(phi)
    z = r*cos(theta)
    
    return x,y,z
```

# Eigen states of X basis

```python
# (1/sqrt(2))(|0> - |1>)

r = 1
theta = pi/4
phi = 0

x,y,z = plot_bloch_vector_sphere(theta, phi, r)
coords=[x,y,z]
plot_bloch_vector(coords)
```

```python
# (1/sqrt(2))(|0> - |1>)

r = 1
theta = -pi/4
phi = 0

x,y,z = plot_bloch_vector_sphere(theta, phi, r)
coords=[x,y,z]
plot_bloch_vector(coords)
```

# Eigen states of Y Basis

```python
# (1/sqrt(2))(|0> + i|1>)

r = 1
theta = pi/4
phi = pi/2

x,y,z = plot_bloch_vector_sphere(theta, phi, r)
coords=[x,y,z]
plot_bloch_vector(coords)
```

```python
# (1/sqrt(2))(|0> - i|1>)
r = 1
theta = pi/4
phi = -pi/2

x,y,z = plot_bloch_vector_sphere(theta, phi, r)
coords=[x,y,z]
plot_bloch_vector(coords)
```
