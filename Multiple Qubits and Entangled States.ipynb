# Multiple Qubits and Entangled States

```python
from qiskit import *
from math import pi
import numpy as np
from qiskit.visualization import *
from qiskit.extensions import Initialize
from matplotlib import style
style.use('dark_background')
```

# Create Entanglement between |00> and |11> states<br>
1. Create superposition in qubit 0.
2. Initialize qubit 0 to 0th state i.e. [1,0]
3. Create entaglement between qubit 0 and 1.

```python
qc = QuantumCircuit(2)
# Apply H-gate to the first:
initial_state=[1, 0] # Sum of the amplitudes must be 1 as it is a pdf
initializer = Initialize(initial_state)
initializer.label = 'init'
qc.append(initializer, [1])
qc.h(0)
qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
final_state = execute(qc,backend).result().get_statevector()
final_state
```

```python
qc.cx(0,1)
qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
final_state = execute(qc,backend).result().get_statevector()
final_state
```

```python
plot_bloch_multivector(final_state)
```

```python
results = execute(qc,backend).result().get_counts()
plot_histogram(results)
```

```python
backend = Aer.get_backend('unitary_simulator')
unitary = execute(qc,backend).result().get_unitary()
unitary
```

```python
qc.measure([0,0], [1,1])
qc.draw()
```
