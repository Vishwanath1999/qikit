## Since we deal here with the X-basis we need to measure them. But in Qiskit we can only measure in X-basis. So we create a function to convert Z to X basis and measure it. At the end we convert it back to z basis. This is done with the helo of Hadmard gate.

```python
def x_measurement(qc, qbit, cbit):
    qc.h(qbit)

    qc.measure(qbit, cbit)

    qc.h(qbit)

    return qbit
```

```python
qc = QuantumCircuit(1,1)
initial_state=[0, 1] # Sum of the amplitudes must be 1 as it is a pdf
initializer = Initialize(initial_state)
initializer.label = 'init'
qc.append(initializer, [0])
```

```python
backend = Aer.get_backend('statevector_simulator')
state = execute(qc, backend).result().get_statevector()
plot_bloch_multivector(state)
```

# The S-gate<br>
The next gate to mention is the S-gate (sometimes known as the  Z -gate), this is an R-gate with  =/2 . It does a quarter-turn around the Bloch sphere. It is important to note that unlike every gate introduced in this chapter so far, the S-gate is not its own inverse! As a result, we will often see the S-gate, (also S-dagger, Sdg or  Z -gate). The S-gate is clearly an R-gate with  =/2 :<br>

The name " Z -gate" is due to the fact that two successively applied S-gates has the same effect as one Z-gate:<br>
ZZ|q> = S|q>
